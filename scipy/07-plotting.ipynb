```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
RESOLUTION = 15
x = np.linspace(0, 2 * np.pi, RESOLUTION)
y = np.sin(x)
```

```python
fig, ax = plt.subplots(1, 1)  # Create a figure containing a single axes.
ax.plot(x, y)  # Plot sin data on the axes.

ax.set_title('Plot')
ax.set_xlabel('radians')
ax.set_ylabel('y')

y = np.cos(x)
ax.plot(x, y)  # Plot cos data on the axes.
```
