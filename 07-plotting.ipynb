```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
dtor = 6.28/360
N = 360
x = np.arange(N)
y = np.ones(N)
for i in range(N):
    y[i] = math.sin(i*dtor)
```

```python
fig, ax = plt.subplots(1, 1)  # Create a figure containing a single axes.
ax.plot(x, y)  # Plot some data on the axes.
ax.set_title('myplot')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
```
